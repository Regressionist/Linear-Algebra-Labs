```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
def sample_points(npoints, mean, std):
    vec = np.random.normal(loc=mean, scale=std, size=(npoints,len(mean)))
    return vec
```

```python
mu = [5,-1]
std=[0.5,2.5]
X=sample_points(50, mu,std)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.xlim([-1,10])
plt.ylim([-8,8])
plt.plot((0,0),(-8,8),'k--',linewidth=0.5)
plt.plot((-1,10),(0,0),'k--',linewidth=0.5)
plt.show()
```

```python
U,sig,Vt = np.linalg.svd(X.T)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(X[:,0],X[:,1])

plt.plot((0,0),(-10,10),'k--',linewidth=0.5)
plt.plot((-3,10),(0,0),'k--',linewidth=0.5)
origin = [0], [0]

plt.arrow(0,0,-2*U[0,0],-2*U[1,0], head_width=.2, color='red')
plt.arrow(0,0,2*U[0,1],2*U[1,1], head_width=.2, color='green')
plt.xlim([-3,10])
plt.ylim([-10,10])
plt.show()
```

### Centering

```python
data = X.T
means = np.matmul(data,np.ones((data.shape[1],1)))/data.shape[1]
data_centered = data - means 
```

```python
plt.figure(figsize=(7,7))
plt.scatter(data_centered[0], data_centered[1])
plt.xlim([-5,5])
plt.ylim([-8,8])
plt.plot((0,0),(-8,8),'k--',linewidth=0.5)
plt.plot((-5,5),(0,0),'k--',linewidth=0.5)
plt.show()
```

```python
U,sig,Vt = np.linalg.svd(data_centered)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(data_centered[0], data_centered[1])

plt.plot((0,0),(-8,8),'k--', linewidth=0.5)
plt.plot((-5,5),(0,0),'k--', linewidth=0.5)

plt.arrow(0,0,-2*U[0,0],-2*U[1,0], width=.05,head_width=.2, color='red')
plt.arrow(0,0,U[0,1],U[1,1], width=.05, head_width=.2, color='green')
plt.xlim([-5,5])
plt.ylim([-8,8])
plt.show()
```

### Scaling 

```python
scaling_factor = np.array([8,1]).reshape(-1,1)
data_new = data_centered * scaling_factor
```

```python
plt.figure(figsize=(7,7))
plt.scatter(data_new[0], data_new[1])
plt.xlim([-12,12])
plt.ylim([-12,12])
plt.plot((0,0),(-12,12),'k--',linewidth=0.5)
plt.plot((-12,12),(0,0),'k--',linewidth=0.5)
plt.show()
```

```python
U,sig,Vt = np.linalg.svd(data_new)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(data_new[0], data_new[1])

plt.plot((0,0),(-12,12),'k--',linewidth=0.5)
plt.plot((-12,12),(0,0),'k--',linewidth=0.5)

plt.arrow(0,0,-3*U[0,0],-3*U[1,0],width=.15, head_width=.5, color='red')
plt.arrow(0,0,1.5*U[0,1],1.5*U[1,1],width=.15, head_width=.5, color='green')
plt.xlim([-12,12])
plt.ylim([-12,12])
plt.show()
```
